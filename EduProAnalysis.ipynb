```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
users = pd.read_excel(
    "EduPro Online Platform.xlsx",
    sheet_name="Users"
)
courses = pd.read_excel(
    "EduPro Online Platform.xlsx",
    sheet_name="Courses"
)
transactions = pd.read_excel(
    "EduPro Online Platform.xlsx",
    sheet_name="Transactions"
)
teachers = pd.read_excel(
    "EduPro Online Platform.xlsx",
    sheet_name="Teachers"
)
users.head()
```

```python
print(users.isnull().sum())

print(courses.isnull().sum())

print(transactions.isnull().sum())
```

```python
users.drop_duplicates(inplace=True)

courses.drop_duplicates(inplace=True)

transactions.drop_duplicates(inplace=True)
```

```python
merged_df = transactions.merge(
    users,
    on="UserID"
)

merged_df = merged_df.merge(
    courses,
    on="CourseID"
)

merged_df.head()
```

```python
bins = [0,18,25,35,45,100]

labels =[
    "<18",
    "18-25",
    "26-35",
    "36-45",
    "45+"
]

merged_df["AgeGroup"] = pd.cut(
    merged_df["Age"],
    bins=bins,
    labels=labels
)

merged_df.head()
```

```python
total_enrollments = merged_df.shape[0]

print("Total Enrollments:")

print(total_enrollments)
```

```python
merged_df["AgeGroup"].value_counts().plot(
    kind="bar"
)

plt.title("Age Group Distribution")

plt.xlabel("Age Groups")

plt.ylabel("Enrollments")

plt.show()
```

```python
merged_df["Gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Gender Distribution")

plt.show()
```

```python
plt.figure(figsize=(14,6))

merged_df["CourseCategory"].value_counts().plot(
    kind="bar"
)

plt.title(
    "Course Category Popularity",
    fontsize=16
)

plt.xlabel(
    "Category",
    fontsize=12
)

plt.ylabel(
    "Enrollments",
    fontsize=12
)

plt.xticks(
    rotation=45,
    ha="right"
)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,6))

heatmap_data = pd.crosstab(
    merged_df["AgeGroup"],
    merged_df["CourseCategory"]
)

sns.heatmap(
    heatmap_data,
    annot=True,
    fmt="d",
    cmap="YlOrRd"
)

plt.title(
    "Age Group vs Course Category"
)

plt.xticks(rotation=45)

plt.show()
```
